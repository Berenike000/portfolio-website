```python
# Q1: Generate NEPTUN code from student name, birth date, and mother's name
def generate_neptun_code(name, birth_date, mother_name):
    import hashlib
    
    # Convert to uppercase and remove spaces
    clean_name = name.upper().replace(" ", "")
    clean_mother_name = mother_name.upper().replace(" ", "")
    
    # Extract year, month, day from birth date
    birth_parts = birth_date.split('/')
    year = birth_parts[0]
    month = birth_parts[1]
    day = birth_parts[2]
    
    # Create a sophisticated hash using SHA-256
    combined_input = f"{clean_name}_{birth_date}_{clean_mother_name}"
    hash_object = hashlib.sha256(combined_input.encode())
    hash_hex = hash_object.hexdigest()
    
    # Convert hash to a mix of letters and numbers
    code = ""
    valid_chars = "ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789"  # 36 possible characters
    
    # Use different parts of the hash for better distribution
    for i in range(6):
        # Take 2 hex characters at a time and convert to decimal
        hex_pair = hash_hex[i*2:(i*2)+2]
        decimal_value = int(hex_pair, 16)
        
        # Map to our valid character set (0-35)
        char_index = decimal_value % len(valid_chars)
        code += valid_chars[char_index]
    
    # Ensure we have some letters by checking the first character
    # If first character is a number, replace it with a letter based on name
    if code[0].isdigit():
        name_hash = sum(ord(char) for char in clean_name)
        letter_index = name_hash % 26
        code = chr(ord('A') + letter_index) + code[1:]
    
    return code[:6]  # Ensure exactly 6 characters

# Test the function with example names, birth dates, and mother's names
print("NEPTUN Code Generation (with birth dates and mother's names)")
print("="*65)

example_data = [
    ("Alice Johnson", "1995/03/15", "Mary Johnson"),
    ("Bob Smith", "1996/07/22", "Susan Smith"),
    ("Charlie Brown", "1994/11/08", "Linda Brown"),
    ("Dani Wilson", "1997/01/30", "Patricia Wilson"),
    ("Eve Davis", "1995/09/12", "Jennifer Davis")
]

for name, birth_date, mother_name in example_data:
    neptun_code = generate_neptun_code(name, birth_date, mother_name)
    print(f"{name:15} ({birth_date}) Mother: {mother_name:15}  {neptun_code}")

print("\n" + "-"*65)
print("Testing uniqueness with same names but different details:")

# Same student name, different birth dates and mothers
test_cases = [
    ("Alice Johnson", "1995/03/15", "Mary Johnson"),
    ("Alice Johnson", "1995/03/16", "Mary Johnson"),  # Different birth date
    ("Alice Johnson", "1995/03/15", "Maria Johnson"), # Different mother's name
    ("Alice Johnson", "1996/03/15", "Susan Smith")    # Different birth date and mother
]

for name, birth_date, mother_name in test_cases:
    neptun_code = generate_neptun_code(name, birth_date, mother_name)
    print(f"{name} ({birth_date}) Mother: {mother_name:15}  {neptun_code}")

print("\n" + "-"*65)
print("Testing deterministic behavior (same input = same output):")
test_name = "Alice Johnson"
test_birth = "1995/03/15"
test_mother = "Mary Johnson"

code1 = generate_neptun_code(test_name, test_birth, test_mother)
code2 = generate_neptun_code(test_name, test_birth, test_mother)

print(f"{test_name} ({test_birth}) Mother: {test_mother}  {code1}")
print(f"{test_name} ({test_birth}) Mother: {test_mother}  {code2}")
print(f"Codes match: {code1 == code2}")

print("\n" + "-"*65)
print("Verifying all codes are exactly 6 characters and uppercase:")
for name, birth_date, mother_name in example_data:
    neptun_code = generate_neptun_code(name, birth_date, mother_name)
    print(f"{neptun_code} - Length: {len(neptun_code)}, All uppercase: {neptun_code.isupper()}")
```

```python
class Student:
    def __init__(self, name, birthday, mom_name):
        """Initialize a Student with name, birthday, and mother's name."""
        self.name = name
        self.birthday = birthday
        self.mom_name = mom_name
        # Generate NEPTUN code using the function from previous task
        self.neptun = generate_neptun_code(name, birthday, mom_name)
    
    def __str__(self):
        """Return a formatted string representation of the student."""
        return f"{self.name} (NEPTUN: {self.neptun}) | born {self.birthday}"


class Class:
    def __init__(self, class_name=""):
        """Initialize a Class with an optional name and empty student list."""
        self.class_name = class_name
        self.students = []
    
    def add_student(self, student):
        """Add a Student object to the class."""
        if isinstance(student, Student):
            self.students.append(student)
        else:
            raise TypeError("Only Student objects can be added to the class")
    
    def add_student_by_data(self, name, birthday, mom_name):
        """Create and add a student using individual data."""
        student = Student(name, birthday, mom_name)
        self.add_student(student)
    
    def get_student_count(self):
        """Return the number of students in the class."""
        return len(self.students)
    
    def list_students(self):
        """Print all students in the class."""
        if not self.students:
            print("No students in the class.")
            return
        
        print(f"Students in {self.class_name if self.class_name else 'the class'}:")
        print("-" * 60)
        for i, student in enumerate(self.students, 1):
            print(f"{i:2d}. {student}")
    
    def find_student_by_neptun(self, neptun_code):
        """Find a student by their NEPTUN code."""
        for student in self.students:
            if student.neptun == neptun_code:
                return student
        return None


# Test the Student and Class classes
print("Testing Student and Class classes:")
print("=" * 60)

# Create a class
my_class = Class("Data Science 101")

# Add students using different methods
print("Adding students to the class...")
my_class.add_student_by_data("Anna Kovacs", "2002/04/18", "Ilona Varga")
my_class.add_student_by_data("Bence Nagy", "2001/11/02", "Maria Toth")
my_class.add_student_by_data("Csilla Toth", "2002/09/25", "Katalin Kiss")

# Create a student object and add it
student4 = Student("Dani Kiss", "2003/02/11", "Zsuzsa Horvath")
my_class.add_student(student4)

print(f"\nClass has {my_class.get_student_count()} students.\n")

# List all students
my_class.list_students()

print("\n" + "=" * 60)
print("Testing student lookup by NEPTUN code:")
# Find a student by NEPTUN code
first_student = my_class.students[0]
found_student = my_class.find_student_by_neptun(first_student.neptun)
if found_student:
    print(f"Found: {found_student}")
else:
    print("Student not found")

```

```python
import pandas as pd
import re

# Q1: Load the CSV and print the first 5 lines
print("Q1: Loading CSV and displaying first 5 rows")
print("=" * 50)
df = pd.read_csv('students_fixed.csv')
print("Students data (head):")
print(df.head())
print(f"\nColumn names: {list(df.columns)}")

# Q2: Inspect types for each column
print("\nQ2: Inspecting data types and missing values")
print("=" * 50)
print("Data types:")
print(df.dtypes)
print("\nMissing values per column:")
print(df.isnull().sum())
print("\nData type analysis:")
print("- name: string")
print("- birthday: date string")
print("- mom_name: string")
print("- exam1, exam2, exam3, exam4: scores (int/float)")

# Q3: Count rows and score columns
print("\nQ3: Counting rows and score columns")
print("=" * 50)
num_students = len(df)
score_columns = [col for col in df.columns if col.startswith('exam')]
num_score_columns = len(score_columns)
print(f"Number of students: {num_students}")
print(f"Number of score columns: {num_score_columns}")
print(f"Score columns: {score_columns}")

# Q4: Validate fields
print("\nQ4: Validating fields")
print("=" * 50)

# Check date format (YYYY/MM/DD or YYYY-MM-DD)
date_pattern = r'^\d{4}[/-]\d{2}[/-]\d{2}$'
invalid_dates = []
for idx, date_str in enumerate(df['birthday']):
    if pd.notna(date_str) and not re.match(date_pattern, str(date_str)):
        invalid_dates.append((idx, date_str))

print(f"Date validation:")
if invalid_dates:
    print(f"Found {len(invalid_dates)} invalid date formats:")
    for idx, date in invalid_dates:
        print(f"  Row {idx}: {date}")
else:
    print("All dates have valid format (YYYY/MM/DD or YYYY-MM-DD)")

# Check scores are 0-100
print(f"\nScore validation (0-100 range):")
invalid_scores = []
for col in score_columns:
    for idx, score in enumerate(df[col]):
        if pd.notna(score) and score != 'absent':
            try:
                score_val = float(score)
                if score_val < 0 or score_val > 100:
                    invalid_scores.append((idx, col, score))
            except ValueError:
                invalid_scores.append((idx, col, score))

if invalid_scores:
    print(f"Found {len(invalid_scores)} invalid scores:")
    for idx, col, score in invalid_scores:
        print(f"  Row {idx}, {col}: {score}")
else:
    print("All scores are in valid range (0-100) or marked as 'absent'")

# Q5: Quick summary
print("\nQ5: Quick summary")
print("=" * 50)
total_missing = df.isnull().sum().sum()
print(f"Number of students: {num_students}")
print(f"Number of exams: {num_score_columns}")
print(f"Total missing entries: {total_missing}")

# Count 'absent' entries separately
absent_count = 0
for col in score_columns:
    absent_count += (df[col] == 'absent').sum()
print(f"Absent entries: {absent_count}")
print(f"Empty entries: {total_missing}")
```
