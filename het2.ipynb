```python
# NEPTUN KD
def f(x):
    return x**2

print(f(5))
import hashlib

def neptun_code(name, birthdate, mothers_name):
    # Combine name, birthdate, and mother's name to ensure uniqueness
    input_str = f"{name}|{birthdate}|{mothers_name}"
    h = hashlib.sha256(input_str.encode('utf-8')).hexdigest()
    # Take the first 6 characters, convert to uppercase, and ensure only letters/digits
    code = ''
    i = 0
    while len(code) < 6:
        c = h[i]
        if c.isalpha() or c.isdigit():
            code += c.upper()
        i += 1
    return code

# Example usage:
example_students = [
    ("Alice Johnson", "2001-05-12", "Mary Johnson"),
    ("Brian Lee", "2000-11-23", "Susan Lee"),
    ("Carmen Diaz", "1999-07-30", "Isabel Diaz"),
    ("David Smith", "2002-03-15", "Linda Smith"),
    ("Emily Zhang", "2001-12-01", "Wei Zhang"),
    ("Frank Perfect", "2000-01-01", "Anna Perfect"),
    # Same name and birthdate, different mother's name
    ("Alice Johnson", "2001-05-12", "Jane Johnson"),
    # Same name, different birthdate and mother's name
    ("Alice Johnson", "2002-08-20", "Mary Johnson"),
]

# 10 more people for testing
more_students = [
    ("George Miller", "1998-04-22", "Helen Miller"),
    ("Hannah Brown", "2003-09-10", "Patricia Brown"),
    ("Ivan Petrov", "2001-01-01", "Olga Petrova"),
    ("Julia Roberts", "1997-06-18", "Betty Roberts"),
    ("Kevin White", "2000-10-05", "Nancy White"),
    ("Laura Green", "2002-02-28", "Diana Green"),
    ("Michael Black", "1999-12-12", "Sarah Black"),
    ("Nina King", "2001-07-07", "Emma King"),
    ("Oscar Turner", "2003-03-03", "Grace Turner"),
    ("Paula Adams", "1998-08-08", "Carol Adams")
]

all_students = example_students + more_students

for name, birthdate, mothers_name in all_students:
    print(f"{name}: {neptun_code(name, birthdate, mothers_name)}")

```
